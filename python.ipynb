```python
for num in range(2, 51):
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num)
```

```python
def prime(n):
    for i in range(2, n):
        if n % i == 0:
            return False
    return n > 1

num = int(input("Enter a number: "))

if prime(num):
    print("Prime")
else:
    print("Not Prime")
```
